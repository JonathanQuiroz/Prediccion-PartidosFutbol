## Instalación

Paso a paso para la instalación:

1. Instalar BeautifulSoup y Requests:
    - pip install beautifulsoup4 Requests
1. Instalar Jupyter Notebook:
    - pip install jupyter[all]
1. Descargar e instalar PhantomJS
    - Para la instalación del PhantomJS dirijase a esta pagina -> [`PhantomJS`](http://phantomjs.org/download.html)
1. Instalar R:
    - apt install -y r-base
1. Clonar repositorio:
    - git clone `https://github.com/JonathanQuiroz/Prediccion-PartidosFutbol.git`

## Consulta

- Para realizar la consulta de información, sobre algun equipo, primero debe importar consulta.py, asi:
    - import consulta as c
- Luego declaramos una variable llamada equipo y le ponemos el nombre de equipo que deseemos:
    - Equipo = "nombre_equipo"
- Por ultimo llamamos la funcion consultaEquipo de consulta.py y la variable equipo como argumento, haciendo decode("utf-8") para el correcto funcionamiento:
    - c.consultaEquipo(Equipo.decode("utf-8"))
    

- Al finalizar, cargara una tabla con los datos de los partidos del equipo requerido:
    - Fecha, torneo, local, GL(goles local), GV(goles visitante), visitante

```python
# Ejemplo
import consulta as c
Equipo = "Albania"
c.consultaEquipo(Equipo.decode("utf-8"))
```
